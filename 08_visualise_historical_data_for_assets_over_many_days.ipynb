```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import json
```

```python
fig_size_x = 10
fig_size_y = 3
```

```python
# 541983310278725782, 541792416037150809, 541480930594258894, 541261957970628703, 541560923639711466

ean = '541560923639711466'

filenames = os.listdir('data_assets')
filenames.sort()

for filename in filenames:

    if not filename.startswith(ean):
        continue

    with open(os.path.join('data_assets', filename)) as f:
        data = json.load(f)
        if len(data['$values']) == 0:
            # print(f"{filename} is empty")
            continue

    # Convert the data to a pandas DataFrame
    df = pd.DataFrame(data['$values'])

    # Convert the 'stateTime' column to datetime
    df['stateTime'] = pd.to_datetime(df['stateTime'])

    # Set 'stateTime' as the index
    df.set_index('stateTime', inplace=True)

    # take the date from the file name
    date = filename.split('.')[0].split('_')[1]
    is_friday = False
    is_weekend = False

    # check if it's in the weekend
    if pd.to_datetime(date).weekday() == 4:
        print(f"{date} is a Friday")
        is_friday = True

    if pd.to_datetime(date).weekday() >= 5:
        # print(f"{date} is a weekend")
        is_weekend = True

    print(f"Plotting {filename}")
    plt.figure(figsize=(fig_size_x, fig_size_y))
    if is_friday:
        # change colour to red
        print("FRIDAY!")
        df['connected'] = df['connected'].astype(int)
        df['connected'].plot(color='red')
    elif is_weekend:
        print("WEEKEND!")
        df['connected'] = df['connected'].astype(int)
        df['connected'].plot(color='green')
    else:    
        df['connected'] = df['connected'].astype(int)
        df['connected'].plot()
    plt.title('Connected Status Over Time')
    plt.xlabel('Time')
    plt.ylabel('Connected Status')
    plt.show()
```
